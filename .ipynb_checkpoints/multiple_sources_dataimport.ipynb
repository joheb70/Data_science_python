```python
import geopy  # Import this to access longitude & latitude of the locatiion with street name, state,country,zipcode in data frame 
dir(geopy)
```

```python
df1["Address"] = df1["Address"]+", "+df1["City"]+ ", "+ df1["Country"]
df1
```

```python
df1
```

```python
from geopy.geocoders import ArcGIS
nom = ArcGIS()
```

```python
nom.geocode("3666 21st St, San Francisco, USA, San Francisco")
```

```python
df1["Coordinates"]=df1["Address"].apply(nom.geocode)
df1
```

```python
df1.Coordinates
```

```python
df1.Coordinates[0]  #To access entire location with latitude and logitude
```

```python
df1.Coordinates[1]
```

```python
df1["Latitude"]=df1["Coordinates"].apply(lambda x: x.latitude) 
df1["Longitude"]=df1["Coordinates"].apply(lambda x: x.longitude) 

 #Create new columns with latitude & longitude  method, lambda is used because 
#-apply method cannot be  applied directly
# -and temporary x variable  is created to apply latitude & longitude method
df1                                                 
```
